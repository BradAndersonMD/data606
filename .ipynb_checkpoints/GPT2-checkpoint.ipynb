```python
import gpt_2_simple as gpt2
from datetime import datetime
```

```python
!nvidia-smi
```

```python
gpt2.download_gpt2(model_name="355M")
```

```python
file_name = r"D:\Users\Brad\Graduate_School\2021\data606\data\final_transcript.txt"
```

```python
sess = gpt2.start_tf_sess()

gpt2.finetune(sess,
              dataset=file_name,
              model_name='355M',
              steps=10000,
              restore_from='fresh',
              run_name='355M',
              print_every=10,
              sample_every=200,
              save_every=500
              )
```

```python
gpt2.generate(sess, run_name="355M", length=250, top_p=.9, nsamples=5, batch_size=5)
```
